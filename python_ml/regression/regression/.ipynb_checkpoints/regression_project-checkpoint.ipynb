## Machine Learning

First of lets see our target variable :

```python
data.SalePrice.hist();
```

Data is skewed to the left, we need the target to be normally distributed, lets log it

```python
data.SalePrice = np.log10(data.SalePrice)
```

```python
data.SalePrice.hist();
```

This looks much better now, let's split our data into training and test sets 

```python
X = data.drop('SalePrice', axis=1)
y = data.SalePrice

X_tr, X_te, y_tr, y_te = train_test_split(X, y, test_size=0.5, random_state=0)
```

Let's first compute the mean absolute error for the baseline

```python
def MAE(y, y_pred):
    return np.mean(np.abs(y-y_pred))
MAE_baseline = MAE(10**y_te,10**np.median(y_tr))
print("The baseline of mean absolute error is :",round(MAE_baseline,2),"$")
```

Let's now look at features we could use to make our first simple model, I'm thinking about looking at the features that are more correlated with the sale price at first, and categorical or ordinal features that have some variations of price depending on the category

```python
simple_cols = data.corr().SalePrice.sort_values(ascending=False)[1:3].index
```

We see that the 2 most important features to determine the sale price are the quality of the house and its ground surface, which seems logical to me so lets work with these and start a model pipeline

```python
X_tr_simple = X_tr[simple_cols]
X_te_simple = X_te[simple_cols]
```

```python
simple_model = Pipeline([
    ("scaler",StandardScaler()),
    ("estimator",Ridge())
])
```

```python
simple_model.fit(X_tr_simple,y_tr);
```

```python
y_pred_simple = simple_model.predict(X_te_simple)
```

Now that we have our predictions, let's compute the MAE.

```python
MAE_simple = MAE(10**y_te,10**y_pred_simple)
print("The mean absolute error of the simple model is :",round(MAE_simple,2),"$")
```

We see that the MAE is better than the baseline so we are on the right track.  
Let's do the same reasoning for the intermediate model, all our data is in numbers now so everything is taken into account for the correlation, so if we pick the 20 highest correlation values they must be the most relevant ones so let's use the first 20 columns 

```python
intermediate_cols = data.corr().SalePrice.sort_values(ascending=False)[1:21].index
```

```python
X_tr_intermediate = X_tr[intermediate_cols]
X_te_intermediate = X_te[intermediate_cols]
```

```python
intermediate_model = Pipeline([
    ("scaler",StandardScaler()),
    ("estimator",Ridge())
])
```

```python
intermediate_model.fit(X_tr_intermediate,y_tr);
```

```python
y_pred_intermediate = intermediate_model.predict(X_te_intermediate)
```

```python
MAE_intermediate = MAE(10**y_te,10**y_pred_intermediate)
print("The mean absolute error of the intermediate model is :",round(MAE_intermediate,2),"$")
```

We see again that the MAE is better, so our model is more performant.  
Now lets use all features in a complex model 


```python
complex_model = Pipeline([
    ("scaler",StandardScaler()),
    ("estimator",Ridge())
])
```

```python
complex_model.fit(X_tr,y_tr);
```

```python
y_pred_complex = complex_model.predict(X_te)
```

```python
MAE_complex = MAE(10**y_te,10**y_pred_complex)
print("The mean absolute error of the complex model is :",round(MAE_complex,2),"$")
```

We see that the MAE is also better as expected, but we could do better if we tune our alpha for the ridge regressor so lets do some regularization

```python
gs_results = []

for alpha in np.logspace(-4, 4, num=100):
    pipe = Pipeline([
        ('scaler', StandardScaler()),
        ('ridge', Ridge(alpha))
    ])
    pipe.fit(X_tr, y_tr)

    gs_results.append({
        'alpha': alpha,
        'train_mae': MAE(10**y_tr, 10**pipe.predict(X_tr)),
        'test_mae': MAE(10**y_te, 10**pipe.predict(X_te)),
    })

gs_results = pd.DataFrame(gs_results)
gs_results.head()
```

```python
plt.semilogx(gs_results['alpha'], gs_results['train_mae'], label='train curve')
plt.semilogx(gs_results['alpha'], gs_results['test_mae'], label='test curve')
plt.xlabel('$log_{10}(alpha)$')
plt.ylabel('MAE')
plt.legend()
plt.show()
```

Ok we see in the graph that there is an optimal value of test MAE around a log10(10^2) alpha, let's tune in this alpha. To do this we take the minimum value of the test curve

```python
optimal_alpha = gs_results.set_index(gs_results.alpha).test_mae.sort_values(ascending=True).index[0]
```

```python
complex_model = Pipeline([
    ("scaler",StandardScaler()),
    ("estimator",Ridge(optimal_alpha))
])
```

```python
complex_model.fit(X_tr,y_tr);
```

```python
y_pred_complex = complex_model.predict(X_te)
```

```python
MAE_complex = MAE(10**y_te,10**y_pred_complex)
print("The mean absolute error of the complex model is :",round(MAE_complex,2),"$")
```

Now let's compare our models to see which one works best :

```python
maes = [MAE_baseline,MAE_simple,MAE_intermediate,MAE_complex]
titles = ["baseline","simple","intermediate","complex"]
xcor = np.arange(len(maes))
plt.bar(xcor,maes)
plt.xticks(xcor,titles)
plt.title("MAE compared to baseline");
```

We see that as expected, our best model is the complex one. Our intermediate model works quite nicely as well.
