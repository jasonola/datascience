# 02. Data exploration

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from PIL import Image
import glob
from sklearn.decomposition import PCA
```

We first load our data

```python
with np.load("train_data.npz",allow_pickle=False) as npztrain:
    train_data = dict(npztrain.items())

with np.load("valid_data.npz",allow_pickle=False) as npzvalid:
    valid_data = dict(npzvalid.items())
    
with np.load("test_data.npz",allow_pickle=False) as npztest:
    test_data = dict(npztest.items())
```

Lets plot a few image from each category, so I make the list of categories by taking the part in the path corresponding with a split.

```python
categories = []
for path in glob.glob("swissroads/train/*/"):
    category = path.split("/")[2]
    categories.append(category)
```

```python
categories
```

Now I iterate 3 times on the categories to open the an image from it, then transforming it to a numpy array and finally plotting the image with its category as a title.

```python
for i in range(3):    
    for category in categories:
        plt.figure()
        pillow_img = Image.open(glob.glob("swissroads/train/*{}/*".format(categories[categories.index(category)]))[i])
        img = np.array(pillow_img)
        plt.imshow(img)
        plt.title(category)

    
```

Let's get our train data

```python
X_train,y_train = train_data.values()
```

let's see the proportion of variance explained, I'll start by showing the first 75 components. We will use PCA and set the number of components to 75.

```python
pca = PCA(n_components=75)
pca.fit(X_train)
pve = pca.explained_variance_ratio_
```

```python
# Create bar plot
xcor = np.arange(1, len(pve) + 1) 
plt.figure(figsize=(20,20))
plt.bar(xcor, pve)
plt.xticks(xcor)

# Add cumulative sum
pve_cumsum = np.cumsum(pve)
plt.step(
    xcor+0.5, 
    pve_cumsum, 
    label='cumulative'
)

# Add labels
plt.xlabel('principal component')
plt.ylabel('proportion of variance explained')
plt.legend()
plt.show()
```

We can see on this scree plot the proportion of variance explained by component. We make a cumsum to add up the proportions in order to see how many of the components explain a certain percentage of the variance. Here we see that the first 2 components already explain 28% of the variance. We achieve 50% with 7 components. Our 75 components achieve around 85%. To have a 100% we would need all of the 281 components. I restrained my scree plot to 75 components for visibility.

Let's now visualise the first 2 components with a scatterplot.

```python
pca2d = PCA(n_components=2)
X_2d = pca2d.fit_transform(X_train)
```

```python
plt.scatter(X_2d[:,0],X_2d[:,1],s=10,c=y_train)
plt.xlabel("First component")
plt.ylabel("second component")
plt.title("Features on a 2d-plot with PCA using the first two components");
```

We can see in this plot that we have 3 main groups that could be separated by a line quite easily. We see 2 clear clusters up and on the right however, there seem to be some confusion in the left cluster as we see in it 3 different classes
