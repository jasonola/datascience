# 05. Logistic regression baseline

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.pipeline import Pipeline
from sklearn.dummy import DummyClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from PIL import Image
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning) 
```

```python
with np.load("train_data.npz",allow_pickle=False) as npztrain:
    train_data = dict(npztrain.items())

with np.load("valid_data.npz",allow_pickle=False) as npzvalid:
    valid_data = dict(npzvalid.items())
    
with np.load("test_data.npz",allow_pickle=False) as npztest:
    test_data = dict(npztest.items())
    
with np.load("raw_train.npz",allow_pickle=False) as npzraw_train:
    raw_train_data = dict(npzraw_train.items())
    
with np.load("raw_test.npz",allow_pickle=False) as npzraw_test:
    raw_test_data = dict(npzraw_test.items())
```

```python
X_train,y_train = train_data.values()
X_valid,y_valid = valid_data.values()
X_test,y_test = test_data.values()
X_train_raw,y_train_raw = raw_train_data.values()
X_test_raw,y_test_raw = raw_test_data.values()
```

Let's compute the baseline for logistic regression. 

```python
dummy = DummyClassifier(strategy="most_frequent")
dummy.fit(X_train,y_train)
print("Baseline accuracy : {:.2f}%".format(100*dummy.score(X_valid,y_valid)))
```

```python
logreg_pipe = Pipeline([
    ('logreg', LogisticRegression(multi_class='ovr', solver='liblinear'))
])
```

```python
logreg_pipe.fit(X_train,y_train)
print("logistic regression accuracy : {:.2f}%".format(100*logreg_pipe.score(X_valid,y_valid)))
```

We see that the accuracy is much better than the baseline. Let's now tune our hyperparameter C to try to get better results.
Let's combine train and validation as the grid search automaticaly splits the train sets in validation

```python
X_train = np.concatenate((X_train,X_valid))
y_train = np.concatenate((y_train,y_valid))
```

```python
grid = {
    'logreg__C': np.logspace(-4, 4, num=10)
}
grid_cv = GridSearchCV(logreg_pipe, grid, cv=5, return_train_score=True)
grid_cv.fit(X_train,y_train);
```

```python
cv_results = pd.DataFrame(grid_cv.cv_results_)
cols = ['mean_test_score', 'param_logreg__C']
cv_results[cols].sort_values('mean_test_score', ascending=False)
```

```python
logreg_pipe_tuned = Pipeline([
    ('logreg', LogisticRegression(multi_class='ovr', solver='liblinear',C=grid_cv.best_params_["logreg__C"]))
])
```

```python
logreg_pipe_tuned.fit(X_train,y_train);
```

```python
fig,axs = plt.subplots(1,10,figsize=(15,9),sharey=True)
probs = []
for i in range(10):
    axs[i].imshow(X_train_raw[i])
    probs.append(np.round(logreg_pipe_tuned.predict_proba(X_train[i].reshape(1,-1)),2))

probs = np.concatenate(probs)
pd.DataFrame(probs,columns=["Bike","Car","Motorcycle","Other","Truck","Van"])
```

Let's now see what happens to the probabilities when we vary the C parameter

```python
fig,axs = plt.subplots(1,10,figsize=(15,9),sharey=True)
probs = []
for c in np.logspace(-4, 4, num=10):
    logreg_pipe_tuned.set_params(logreg__C=c)
    for i in range(10):
        logreg_pipe_tuned.fit(X_train,y_train)
        axs[i].imshow(X_train_raw[i])
        probs.append(np.round(logreg_pipe_tuned.predict_proba(X_train[i].reshape(1,-1)),2))

probs = np.concatenate(probs)
pd.DataFrame(probs,columns=["Bike","Car","Motorcycle","Other","Truck","Van"]).tail(20)
```

```python
pd.DataFrame(probs,columns=["Bike","Car","Motorcycle","Other","Truck","Van"]).head(40)
```

So I varied the regularization strength an I see that the highest probability is still on the same category. Also I see that somehow, as I set the parameters of the regularization strength, my probabilities tend to either 0 or 1. I'm assuming it is learning through the for loop.

Let's compute the accuracy of the model on test data

```python
logreg_pipe_tuned = Pipeline([
    ('logreg', LogisticRegression(multi_class='ovr', solver='liblinear',C=grid_cv.best_params_["logreg__C"]))
])
```

```python
logreg_pipe_tuned.fit(X_train,y_train);
```

```python
logreg_test_acc = 100*logreg_pipe_tuned.score(X_test,y_test)
```

```python
print("Test accuracy : {:.1f}%".format(logreg_test_acc))
```
