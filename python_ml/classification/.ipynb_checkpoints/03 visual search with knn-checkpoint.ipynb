# 03. Visual search with K-nn

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.neighbors import KNeighborsClassifier,NearestNeighbors
from sklearn.pipeline import Pipeline
```

```python
with np.load("train_data.npz",allow_pickle=False) as npztrain:
    train_data = dict(npztrain.items())

with np.load("valid_data.npz",allow_pickle=False) as npzvalid:
    valid_data = dict(npzvalid.items())
    
with np.load("test_data.npz",allow_pickle=False) as npztest:
    test_data = dict(npztest.items())
    
with np.load("raw_train.npz",allow_pickle=False) as npzraw_train:
    raw_train_data = dict(npzraw_train.items())
    
with np.load("raw_test.npz",allow_pickle=False) as npzraw_test:
    raw_test_data = dict(npzraw_test.items())
```

```python
X_train,y_train = train_data.values()
X_valid,y_valid = valid_data.values()
X_test,y_test = test_data.values()
X_train_raw,y_train_raw = raw_train_data.values()
X_test_raw,y_test_raw = raw_test_data.values()
```

Let's create a pipeline for the knn classifier and get the accuracy.

```python
knn_pipe = Pipeline([
    ('knn', KNeighborsClassifier(n_neighbors=10))
])

knn_pipe.fit(X_train, y_train)

accuracy = knn_pipe.score(X_valid, y_valid)
print('Accuracy: {:.1f}%'.format(100*accuracy)) 
```

We see that we get an accuracy of 92.8% with a k-value of 10. Let's tune it with grid search.

```python
k_values = np.r_[1, np.arange(5, 101, step=5)]
```

```python
gs_results = []

for k in k_values:
    knn_pipe.set_params(knn__n_neighbors=k)
    knn_pipe.fit(X_train, y_train)

    gs_results.append({
        'k': k,
        'train_accuracy': knn_pipe.score(X_train, y_train),
        'validation_accuracy': knn_pipe.score(X_valid, y_valid)
    })

gs_results = pd.DataFrame(gs_results)
gs_results.sort_values(by='validation_accuracy', ascending=False).head()
```

We see that we get the best results between 1 and 20 so let's fine tune it

```python
k_values = np.arange(1,20 , step=1)
```

```python
gs_results = []

for k in k_values:
    knn_pipe.set_params(knn__n_neighbors=k)
    knn_pipe.fit(X_train, y_train)

    gs_results.append({
        'k': k,
        'train_accuracy': knn_pipe.score(X_train, y_train),
        'validation_accuracy': knn_pipe.score(X_valid, y_valid)
    })

gs_results = pd.DataFrame(gs_results)
gs_results.sort_values(by='validation_accuracy', ascending=False).head()
```

Now let's pick an image from the test set and plot its 10 nearest neighbors from the train set, we reshape it to use it as a single sample. Then we fit the train set and finaly we pass the sample into the kneighbors method to get the distances and the indices of the nearest neighbors of our test image from the train set.

```python
arbitrary_number = 4
```

```python
X_img = X_test[arbitrary_number].reshape(1,-1)
```

```python
X_img_raw = X_test_raw[arbitrary_number]
```

```python
nn = NearestNeighbors(n_neighbors=10)
```

```python
nn.fit(X_train);
```

```python
indices = nn.kneighbors(X_img)[1][0]
distances = nn.kneighbors(X_img)[0][0]
```

Now creating the dictionary to iterate through both index and distance at the same time. We first plot the image and then we plot its 10 nearest neighbors.

```python
neighbors = {distances[i]: indices[i] for i in range(len(indices))} 
```

```python
plt.imshow(X_img_raw.reshape(299,299,3))
plt.title("Test image")
for dist,idx in neighbors.items():
    plt.figure()
    plt.imshow(X_train_raw[idx])
    plt.title("d = {:.3f}".format(dist))
```

Let's run our pipe on the test set and get the test accuracy

```python
knn_pipe.set_params(knn__n_neighbors=3)
knn_pipe.fit(X_train, y_train)
knn_test_acc = 100*knn_pipe.score(X_test,y_test)
print("Test accuracy : {:.1f}%".format(knn_test_acc))
```
