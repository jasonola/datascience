```python
from pyAudioAnalysis import MidTermFeatures as aF
import os
import numpy as np
import glob
import pandas as pd
```

```python
#dirlist = glob.glob("genres/*") not good
```

```python
# select number of classes for testing purposes because extraction takes long time

num_classes = 10 # max 10
```

```python
#getting the paths of directories

dirlist2 = ["genres/"+genre for genre in next(os.walk('genres/'))[1]]
dirlist2
```

```python
#keeping the class names

class_names = [os.path.basename(d) for d in dirlist2]
class_names
```

```python
#Setting midterm window, midterm step, shortterm window and shortterm step
#these are just variables for parameters in next cell's function

m_win, m_step, s_win, s_step = 1, 1, 0.1, 0.05 
```

```python
# Extract the features, here I only extracted features for the first 2 genres in our dirlist2 
# for test purposes because extracting features takes much time. Doing this for the 10 classes takes about 15 mins.

features = aF.multiple_directory_feature_extraction(dirlist2[:num_classes],m_win,m_step,s_win,s_step)
```

```python
# we see that we have 2 sets of arrays corresponding to each of our 2 classes

features
```

```python
# in the first index of features we have the arrays of features for our 2 classes

features[0]
```

```python
# here we have the 136 features for our first class pop with the 100 lines corresponding to the wave file

pd.DataFrame(features[0][0])
```

```python
#here we stack our 2 classes 

both_classes = []
for i in range(num_classes):
    both_classes.append(features[0][i])
    
full = np.concatenate(both_classes)

full.shape
```

```python
# Transform to dataframe for both classes

df = pd.DataFrame(full)
df
```

```python
# we need to add class names to df now , first we create the classes 

# classes = np.concatenate([[class_names[0]] * 100, [class_names[1]] * 100])
```

```python
classes = np.concatenate([[class_name]*100 for class_name in class_names[:num_classes]])
classes
```

```python
# then we add it to our dataframe

df["classes"] = classes
df
```

```python
df.to_csv("raw_features.csv")
```
