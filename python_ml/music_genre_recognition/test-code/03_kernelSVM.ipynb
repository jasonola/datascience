```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from yellowbrick.classifier import ConfusionMatrix, ClassificationReport
import pickle
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.decomposition import PCA
plt.style.use("default")
```

```python
data = pd.read_csv("preprocessed_features.csv",index_col="Unnamed: 0")
```

```python
X = data.drop("classes", axis = 1)
y = data.classes
```

```python
X_tr, X_te, y_tr, y_te = train_test_split(X, y, train_size=0.8, stratify=y)
```

```python
kernel_svm = Pipeline([
    ("kernel_svm",SVC())
])
grid_kernel = {
    "kernel_svm__C":np.logspace(-4,4,num=9),
    "kernel_svm__gamma":np.logspace(-4,4,num=9)
}
```

```python
gs_kernel_svm = GridSearchCV(kernel_svm,grid_kernel,cv=5,return_train_score=True)
gs_kernel_svm.fit(X_tr,y_tr);
```

```python
cv_results = pd.DataFrame(gs_kernel_svm.cv_results_)[["mean_test_score","param_kernel_svm__C","param_kernel_svm__gamma"]]
cv_results.sort_values(by="mean_test_score",ascending=False).head(3)
```

```python
cm_svm= ConfusionMatrix(gs_kernel_svm)
cm_svm.fit(X_tr,y_tr)
cm_svm.score(X_te,y_te)
cm_svm.show();
```

```python
report = ClassificationReport(gs_kernel_svm)
report.fit(X_tr,y_tr)
report.score(X_te,y_te)
report.show();
```

```python
pca_kernel_svm = Pipeline([
    ("pca",PCA(n_components=57)),
    ("kernel_svm",SVC(gamma=0.01, C=10000))
])

pca_kernel_svm.fit(X_tr, y_tr)
```

```python
pca_kernel_svm.score(X_te,y_te)
```

```python
kernel_pca_cm = ConfusionMatrix(pca_kernel_svm)
kernel_pca_cm.score(X_te,y_te)
kernel_pca_cm.show();
```

```python
report = ClassificationReport(pca_kernel_svm)
report.score(X_te,y_te)
report.show();
```

```python
# we keep the grid search without pca as it shows better overall results

pickle.dump(gs_kernel_svm, open("svm.sav","wb"))
```
